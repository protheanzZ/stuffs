```python
def process_21(df):   
    is50 = df.loc[4]
    
    rec_100 = df.loc[8:10].dropna().values.ravel()
    rec_10 = df.loc[11:13].dropna().values.ravel()
    rec_50 = df.loc[14:16].dropna().values.ravel()

    urine_blank = df.loc[17:19].dropna().values.ravel()

    matrix_effect = df.loc[20:22].dropna().values.ravel()
    pure_water = df.loc[23:25].dropna().values.ravel()

    results = pd.DataFrame()

    exps = [rec_10, rec_50, rec_100]
    theo_value = np.array([10, 50, 100])

    ub_mean = urine_blank.mean()
    if not urine_blank.any():
        ub_mean = 0

    for i, e in enumerate(exps):
        rec_set = (e-ub_mean)/theo_value[i]
        mean = rec_set.mean()
        std = rec_set.std()

        rec_set = pd.DataFrame(rec_set).T
        rec_set['mean'] = mean
        rec_set['std'] = std
        results = results.append(rec_set)

    results.index = ['%s_10' % compound, '%s_50' % compound, '%s_100' % compound]
    results['urine_base'] = ub_mean - pure_water.mean()
    
    me = pd.DataFrame(matrix_effect / is50).T.rename({0:'ME-1', 1:'ME-2', 2:'ME-3'}, axis=1)
    me['ME_mean'] = (matrix_effect / is50).mean()
    me['ME_std'] = (matrix_effect / is50).std()
    me.index = [compound]
    
    return results, me
```

```python
re_total_21 = pd.DataFrame()
me_total_21 = pd.DataFrame()

for i in range(100):
    try:
        df = pd.read_excel('(COPY)PFCs-21-xwp-20190122-2-正确定量.xlsx', i).rename({'Theoretical Amt': 'ta', 'Calculated Amt': 'ca'}, axis=1)
        if np.all(df['ta'] == 1):
            continue
        compound = df['Compound'][0]
        df['ca'] = df.apply(deal_na, axis=1)
        df = df['ca']
    except IndexError:
        print('bound reached')
        break
        
    rr, me = process_21(df)
    re_total_21 = re_total_21.append(rr)
    me_total_21 = me_total_21.append(me)
```
